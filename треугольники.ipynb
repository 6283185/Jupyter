```python
s = '*'
p = ' '
i = 1
while i <= 7:
  print(p*((7-i)//2)+s*i)
  i += 2
```

```python
n=int(input())
c=1
while c<=n:
    print (' '*((n-c)//2), '*'*c)
    c+=2
```

```python
index = int(input("Насколько большой треугольник делать? (введи число)"))
s = '*'
s_i = 1

jump = ' '
jump_i = index

for i in range(0, index):
    print(jump * jump_i + s * s_i)
    s_i += 2
    jump_i -= 1

print("Happy New Year, you, python-guy!")
```
